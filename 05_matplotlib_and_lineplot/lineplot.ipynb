# Data Visualization

At this point, we have learned how to handle data using Pandas. Now, let's start visualizing the data! YES! Let's start making plots! Finally!

But first, what is data visualization?

*Data visualization is the graphical representation of data and information. It involves the process of creating visual representations of data.*

## Plot Libraries

Basic charts and graphs are the simplest forms of data visualization to represent numerical data. In Python, we have several plotting libraries that make our lives easier when we need to generate the graphical representation of data and information. In this class, we will explore one of these libraries: the **Matplotlib** library!

### Matplotlib

- Created by John Hunter, an American neurobiologist.
- Matplotlib is a general-purpose plotting library that offers a wide variety of plots and customization options.
- It integrates well with other libraries and frameworks, allowing the combination of the strengths of different libraries.

### Architecture

**Scripting Layer (pyplot)**

The Scripting Layer, provided by `pyplot`, facilitates the creation of common types of graphs and charts with syntax similar to MATLAB. Designed to be user-friendly, it is often used for quick and straightforward data visualization without the need for in-depth customization. This layer is ideal for those new to Matplotlib or in need of rapid visualization creation.

**Artist Layer (Artist)**

The Artist Layer is where much of the detailed work occurs. It offers classes that allow detailed control over the plot's appearance, such as lines, text, and shapes. This layer enables users to customize nearly every element of a graph or chart, making it powerful for crafting complex visualizations. Users can interact with and modify the artists in a figure directly.

There are two types of **Artist** objects:
1. **Primitive**: Line2D, Rectangle, Circle, and Text.
2. **Composite**: Axis, Tick, Axes, and Figure.

*Attention: The top-level Matplotlib object containing and managing all the elements in a graphic is the Figure artist. The most crucial composite artist is the Axis because it's where most of the Matplotlib API plotting methods are defined.*

**Backend Layer (FigureCanvas, Renderer, Event)**

The Backend Layer addresses figure rendering and event handling. This layer comprises three main components:
- **FigureCanvas**: The area onto which the figure is drawn.
- **Renderer**: The engine operating behind the scenes to create visual representations of the plots. It translates the higher-level objects from the Artist Layer into visual elements.
- **Event**: The mechanism that handles user inputs, such as keyboard strokes and mouse clicks.

### The Line Plot

The line plot displays data as a series of data points connected by straight lines.

**Use cases:**

- Displays trends over time

- Compares data sets with a continuous independent variable (like age or time)

# Graph Aesthetics

There are several parameters we can adjust using Matplotlib to make your graphs more visually appealing and engaging.

### Graph Size

The first thing to consider is the size of the graph. We can define the size of the graph by using `plt.figure(figsize=(width, height))`. This needs to be the first thing we define in plt, before any other operations.


### Titles & Labels

We can also add a title to your graph and labels to the axes:

```python
plt.title("Title Here")
plt.ylabel("Y-Axis Label")
plt.xlabel("X-Axis Label")
```

Besides the text, optionally, we can define the font size by passing the argument `fontsize`: `plt.title("Title Here", fontsize=12)`. This argument also works for the labels.

*Note that it is common practice to add the title and the labels just after calling the `plt.plot` function.*


### Line Style, Markers, and Colors

In Matplotlib, customizing the look of your plots with line styles, markers, and colors enhances clarity and visual appeal.

- **Line Style**: This alters the appearance of the plot line. For example, using `'-.'` results in a dash-dot line, while `'--'` gives you a dashed line. These styles help differentiate between different data series or emphasize particular trends.

  Example: `plt.plot(x, y, linestyle='-.')`

- **Markers**: These symbols mark your data points on the graph. For instance, `'o'` denotes circle markers, and `'x'` marks each point with an x. Markers are great for highlighting specific data points or for plots where individual data points are as significant as the trend they form.

  Example: `plt.plot(x, y, marker='o')`

- **Colors**: Color choice can make your data stand out or convey additional meanings (like using red for losses and green for gains). In Matplotlib, you can specify colors in many ways, such as by name (`'red'`), hexadecimal code (`'#FF0000'`), or RGB tuple (`(1, 0, 0)`).

  Example: `plt.plot(x, y, color='red')`


### Changing the tickes

Sometimes we need to change the size and even the rotation of the tickes x-axis or y-axis. In matplotlib we can do that by using the following calls:

```python
plt.xticks(rotation=45)  # Rotates the x-axis ticks by 45 degrees
plt.yticks(fontsize=8)  # Changes the font size of the y-axis ticks to 8
```


### Grid

Sometimes it's not clear for what value of x and y a certain point belogs. Matplotlib allow us to make the graph mor readble by add a background grid.
To add the grid you need to call `plt.grid()`

### Legend

We can also add a legend box to our graph. This is particularly useful when we have more than one line in the graph. Matplotlib allows us to easily add it by including a label in the `plot` call:

```python
plt.plot(x, y, label="a label")
plt.legend()
```

### Defining the Axis Limits and Frequency

Sometimes, it is necessary to specify the minimum and maximum values displayed on the axis. For instance, when you want to make a y-axis starting from 0 or when we want to make two distinct graphs have the same range of values to compare them.

In Matplotlib, you can define the limits of the axes using the following calls:

```python
# Defining the limits of the axes
plt.xlim(0, 9)  # Sets the limits of the x-axis from 0 to 9
plt.ylim(0, 81)  # Sets the limits of the y-axis from 0 to 81
```

Another important aspect of the axes is their frequency. By default, Matplotlib will add a frequency that best fits the graph. However, sometimes we want to see different points. For instance, instead of a frequency of 10 points, we might want it to display every 5 points. To change the frequency of the ticks, we need to use the `matplotlib.ticker` module, which provides a variety of tick locators. For adjusting the tick frequency, the `MultipleLocator` class allows you to set a tick on the x-axis for every fixed interval that you specify:

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker # <-- Need to import this

# Set the frequency of ticks on the x-axis to every 10 units
ax = plt.gca()  # Get current axis
ax.xaxis.set_major_locator(ticker.MultipleLocator(10))
```

### Saving the Graph

In the end, we can save the graph to a file. This should be the last call in your code and usually replaces the `plt.show()` call.

To save a file using matplotlib, we use the following call:

```python
plt.savefig("file_name.png", dpi=300, bbox_inches='tight')
```

The `dpi` variable is used to specify the resolution of the file in dots per inch. A higher `dpi` value results in a higher resolution image, which is particularly useful for publications or presentations where clarity is essential.

The `bbox_inches='tight'` parameter ensures that the bounding box of the figure is as tight as possible. This helps to minimize any excess whitespace around the figure, making the saved image more compact and focused on the graph itself.

Note that we can save files in different formats, such as PDF, by simply changing the extension in the file name (e.g., `file_name.pdf`).

# Activity 03

Using the dataset `Canada.csv`, generate a line plot by continent (column `AreaName`).

Through the graph, identify the continent with the highest migration over the years and generate a line plot featuring countries from this continent. Since it includes many countries, create the plot with only the countries whose total migration over the years exceeds 60,000.

*Tip: Creating a 'total' column may facilitate the generation of your graph.*

Which continent and country have the highest migrations to Canada?

````python
# Do the Activity 03 Here

# Create a cleaner dataframe and keep the year columns together.
columns = {"OdName": "Country",
           "AreaName": "Region"}

df_activity03 = df.rename(columns=columns).drop(["Type", "Coverage", "AREA", "REG", "DEV", "DevName"], axis=1)

x_axis = [str(x) for x in range(1980, 2014)]
df_activity03["Total"] = df_activity03[x_axis].sum(axis=1)

# Generate a line plot by continent.
df_continents = df_activity03.groupby("Region")[x_axis].sum()
df_continents["Total"] = df_continents[x_axis].sum(axis=1)

plt.figure(figsize=(50, 25), facecolor="#111827")
ax = plt.gca()
ax.set_facecolor("#F9FAFB")

colors = ["#2563EB", "#DC2626", "#16A34A", "#9333EA", "#EA580C", "#0891B2"]

for region, color in zip(df_continents.index, colors):
    plt.plot(x_axis,
             df_continents.loc[region, x_axis],
             marker="o",
             linestyle="--",
             linewidth=3,
             markersize=10,
             color=color,
             label=region)

plt.title("Immigration to Canada by Continent", fontsize=50, color="white")
plt.xlabel("Years", fontsize=40, color="white")
plt.ylabel("Number of immigrants", fontsize=40, color="white")

plt.xticks(rotation=45, fontsize=30, color="white")
plt.yticks(fontsize=30, color="white")
plt.grid(True, color="#9CA3AF", alpha=0.5)
plt.legend(fontsize=28)

for spine in ax.spines.values():
    spine.set_color("#111827")

plt.show()

# Identify the continent with the highest total immigration.
top_continent = df_continents["Total"].idxmax()
print("Continent with the highest migration:", top_continent)

# Plot only countries from the top continent with total migration above 60,000.
df_top_continent = df_activity03[(df_activity03["Region"] == top_continent) & (df_activity03["Total"] > 60000)]
df_top_continent = df_top_continent.sort_values("Total", ascending=False)

plt.figure(figsize=(50, 25), facecolor="#0F172A")
ax = plt.gca()
ax.set_facecolor("#F8FAFC")

for index, row in df_top_continent.iterrows():
    plt.plot(x_axis,
             row[x_axis],
             marker="o",
             linestyle="--",
             linewidth=3,
             markersize=9,
             label=row["Country"])

plt.title("Asian Countries with More Than 60,000 Immigrants to Canada", fontsize=50, color="white")
plt.xlabel("Years", fontsize=40, color="white")
plt.ylabel("Number of immigrants", fontsize=40, color="white")

plt.xticks(rotation=45, fontsize=30, color="white")
plt.yticks(fontsize=30, color="white")
plt.grid(True, color="#94A3B8", alpha=0.5)
plt.legend(fontsize=24, ncol=2)

for spine in ax.spines.values():
    spine.set_color("#0F172A")

plt.show()

top_country = df_top_continent.iloc[0]
print("Country with the highest migration:", top_country["Country"])
print("Total migration:", top_country["Total"])
````
