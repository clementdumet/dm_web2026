## Exercise 1

Read all the files and create ONE pandas DataFrame containing the full dataset. (The full
dataset is supposed to have 12 columns and 8800 rows.)


```python
import pandas as pd


df_g = pd.read_csv("dataset/netflix_titles_G.csv")
df_nc = pd.read_csv("dataset/netflix_titles_NC-17.csv")
df_nr = pd.read_csv("dataset/netflix_titles_NR.csv")
df_pg13 = pd.read_csv("dataset/netflix_titles_PG-13.csv")
df_pg = pd.read_csv("dataset/netflix_titles_PG.csv")
df_r = pd.read_csv("dataset/netflix_titles_R.csv")
df_tv14 = pd.read_csv("dataset/netflix_titles_TV-14.csv")
df_tvg = pd.read_csv("dataset/netflix_titles_TV-G.csv")
df_tvma = pd.read_csv("dataset/netflix_titles_TV-MA.csv")
df_tvpg = pd.read_csv("dataset/netflix_titles_TV-PG.csv")
df_tvy = pd.read_csv("dataset/netflix_titles_TV-Y.csv")
df_tvy7fv = pd.read_csv("dataset/netflix_titles_TV-Y7-FV.csv")
df_tvy7 = pd.read_csv("dataset/netflix_titles_TV-Y7.csv")
df_ur = pd.read_csv("dataset/netflix_titles_UR.csv")

df = pd.concat([df_g, df_nc, df_nr, df_pg13, df_pg, df_r, df_tv14, df_tvg,
                df_tvma, df_tvpg, df_tvy, df_tvy7fv, df_tvy7, df_ur], ignore_index=True)
df.info()
```

## Exercise 2 - Data Cleaning

```python
# a.​ Check for null values in your data. What are the top TWO columns with the most null values?
df.isnull().sum()

# director : 2631 null values
# country : 830 null values
# cast : 825 null values
```

```python
# b. Remove the first and the third column with the most null values from your data using the drop procedure.

columns_to_drop = ['director', 'cast']
df = df.drop(columns=columns_to_drop)
df.head()
```

```python
# c.  The country column has the second most null values. Handle it: Replace all the NaN values in the country column with "United States", as Netflix was created inthe USA and every show is aired on Netflix US.

df['country'] = df['country'].fillna('United States')
df.isnull().sum()
```

```python
# c.  Since we already have the release year for each movie, we can drop the date_added column. Drop the date_added column.

df = df.drop(columns=['date_added'])
df.info()
```

## Exercise 5
Considering the full dataset, what is the percentage of releases that are listed as “Action
& Adventure”?


```python
total_entries = len(df)
df_action_adventure = df[df['listed_in'] == 'Action & Adventure']
total_action_adventure = len(df_action_adventure)

print(f"Total entries: {total_entries}")
print(f"Total entries in Action & Adventure: {total_action_adventure}")
print(f"Percentage of Action & Adventure: {total_action_adventure / total_entries * 100:.2f}%")
```

Considering the full dataset, what are the percentages of movies and TV shows?

## Exercise 12 - Movies Analysis
Create a dataframe with only the Movies and answer the following questions:


```python
df_movies = df[df['type'] == "Movie"]
```

```python
# a.​ How many rows does the Movies dataframe have?
print(f"The Movies dataframe has {len(df_movies)} rows.")
```

```python
# b. Considering the total number of MOVIES in the dataset, what is the percentage of movies in each rating?
gb = df_movies.groupby('rating')

for name, group in gb:
    percentage = len(group) / len(df_movies) * 100
    print(f"Rating {name}: {percentage:.2f}%")
```

```python
# C.  What are the top 10 genres of movies?

df_movies['listed_in'].value_counts().head(10)
```

```python
# D. What is the year when the most MOVIES were released?

df_movies['release_year'].value_counts().head(1)
```

```python
# e; Print the top 15 oldest movies available on Netflix.

df_movies.sort_values(by='release_year', ascending=True).head(15)[['title', 'release_year', 'listed_in']]
```
