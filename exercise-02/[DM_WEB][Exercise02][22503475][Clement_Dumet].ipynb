## 1. Build Full DataFrame

Read all 14 CSV files from `exercise-02/dataset` and create **one** pandas DataFrame containing the full dataset (expected shape: 8800 rows, 12 columns).


```python
from pathlib import Path

data_dir = Path("dataset")
csv_files = sorted(data_dir.glob("*.csv"))

df_full = pd.concat([pd.read_csv(file) for file in csv_files], ignore_index=True)
print("Shape:", df_full.shape)
df_full.head()
```

## 2. Data Cleaning

Perform the cleaning procedures:
- (a) Check null values and identify the top 2 columns with most nulls.
- (b) Remove the 1st and 3rd columns with most null values using `drop`.
- (c) Replace all `NaN` in `country` with `"United States"`.
- (d) Drop `date_added` column.


```python
# Write your solution here

#(a)
null_count = df_full.isnull().sum()
sorted_null = null_count.sort_values(ascending=False)
print(sorted_null.head(2))

#(b)
to_drop = [sorted_null.index[0], sorted_null.index[2]]
df_full.drop(columns = to_drop, inplace=True)

#(c)
df_full['country'] = df_full['country'].fillna('United States')

#(d)
df_full.drop(columns = ['date_added'], inplace=True)


df_full.head()
```

## 12. Movies-Only DataFrame

Create a Movies-only DataFrame and answer:
- (a) Number of rows
- (b) Percentage of movies in each rating
- (c) Top 10 movie genres
- (d) Year with most movie releases
- (e) Top 15 oldest movies available on Netflix


```python
# Write your solution here
# (a)
df_movies = df_full[df_full['type'] == 'Movie']
print(f"num of movies: {len(df_movies)}")

# (b)
percent_ratings = ((df_movies["rating"].value_counts(normalize=True) * 100).round(2).astype(str)+ " %")
print(percent_ratings )

# (c)
genre_count = df_movies['listed_in'].value_counts()
print(genre_count.head(10))

# (d)
year_count = df_movies['release_year'].value_counts()
print(year_count.head(10))

# (e)
oldest_movies = df_movies.sort_values('release_year')
print(oldest_movies.head(15))
```
