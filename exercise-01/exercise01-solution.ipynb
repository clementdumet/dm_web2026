### 9. Convert Fahrenheit to Celsius
Write a program that asks for the temperature in degrees Fahrenheit, converts, and displays the temperature in degrees Celsius.  
Use the formula:  
**C = 5 * ((F - 32) / 9)**

````python
def fahrenheit_to_celsius(fahrenheit):
    return 5 * ((fahrenheit - 32) / 9)

fahrenheit = input("Enter the temperature in Fahrenheit: ")

try:
    fahrenheit = float(fahrenheit)
    celsius = fahrenheit_to_celsius(fahrenheit)
    print(f"{fahrenheit}°F is equal to {celsius}°C")
except ValueError:
    print("Invalid input. Please enter a valid number.")
````

### 10. Convert Celsius to Fahrenheit
Write a program that asks for the temperature in degrees Celsius, converts, and displays it in degrees Fahrenheit.

### 14. Fish Weight Regulation
João, an honest fisherman, bought a microcomputer to monitor his daily work yield.  
According to São Paulo's fishing regulations, any fish weight **above 50 kg** requires a fine of **R$ 4.00 per excess kilogram**.  
Write a program that:
- Reads the weight of the fish.
- Calculates and stores the excess weight (if any).
- Computes the fine amount João must pay.
- Prints the results with appropriate messages.


````python
fish_weight = input("Enter the weight of the fish (in kg): ")

try:
    fish_weight = float(fish_weight)
    limit = 50.0
    fine_per_kg = 4.0

    if fish_weight > limit:
        excess = fish_weight - limit
        fine = excess * fine_per_kg
    else:
        excess = 0.0
        fine = 0.0

    
    print(f"Total fish weight: {fish_weight:.2f} kg")
    print(f"Excess weight: {excess:.2f} kg")
    print(f"Fine to be paid: R$ {fine:.2f}")

except ValueError:
    print("Invalid input. Please enter a valid number for the weight.")
````

### 15. Salary Calculation with Deductions
Write a program that asks how much you earn per hour and the number of hours worked in the month.  
Then, calculate and display:
- **Gross salary**.
- **INSS tax (8%)**.
- **Union tax (5%)**.
- **Income tax (11%)**.
- **Net salary** after deductions.

**Formula:**
```
+ Gross Salary : R$
- IR (11%) : R$
- INSS (8%) : R$
- Union (5%) : R$
= Net Salary : R$
```
**Note:**  
Net Salary = Gross Salary - Deductions.

````python
earn_per_hour = input("Enter your earnings per hour: ")
hours_worked_in_a_month = input("Enter the number of hours worked in a month: ")

#- **Gross salary**.
#- **INSS tax (8%)**.
#- **Union tax (5%)**.
#- **Income tax (11%)**.
#- **Net salary** after deductions.


try:
    earn_per_hour = float(earn_per_hour)
    hours_worked_in_a_month = int(hours_worked_in_a_month)
    gross_salary = earn_per_hour * hours_worked_in_a_month
    inss_tax = gross_salary * 0.08
    union_tax = gross_salary * 0.05
    income_tax = gross_salary * 0.11
    net_salary = gross_salary - (inss_tax + union_tax + income_tax)
    print(f"+ Gross Salary: R$ {gross_salary:.2f}")
    print(f"- INSS Tax (8%): R$ {inss_tax:.2f}")
    print(f"- Union Tax (5%): R$ {union_tax:.2f}")
    print(f"- Income Tax (11%): R$ {income_tax:.2f}")
    print(f"= Net Salary: R$ {net_salary:.2f}")
except ValueError:
    print("Invalid input. Please enter valid numbers for earnings and hours worked.")



    
````

### 16. Paint Shop - Buying Large Cans
Write a program for a paint shop. The program should ask for the **area (in square meters)** to be painted.  
Consider:
- 1 liter of paint covers **3 square meters**.
- Paint is sold in **18-liter cans**, costing **R$ 80.00** each.
- Display the **number of cans required** and **total price**.


````python
import math

area_to_paint_in_m2 = input("Enter the area to paint in m²: ")
liters_per_can = 18
can_price = 80.00

try:
    area_to_paint_in_m2 = float(area_to_paint_in_m2)
    liters_needed = area_to_paint_in_m2 / 3
    cans_needed = math.ceil(liters_needed / liters_per_can)
    total_price = cans_needed * can_price
    print(f"Total number of cans needed for {area_to_paint_in_m2} m²: {cans_needed}")
    print(f"Total price: R$ {total_price:.2f}")
except ValueError:
    print("Invalid input. Please enter a valid number for the area to paint.")
````

### 17. Paint Shop - Optimized Buying
Modify the previous program to allow different purchase options.  
Consider:
- 1 liter of paint covers **6 square meters**.
- Paint is sold in:
  - **18-liter cans** for **R$ 80.00**.
  - **3.6-liter gallons** for **R$ 25.00**.
- Calculate and display:
  1. The cost if buying **only 18-liter cans**.
  2. The cost if buying **only 3.6-liter gallons**.
  3. The **optimal mix** of cans and gallons to minimize waste (with a 10% safety margin).

````python
import math

LITERS_PER_M2 = 1 / 6  # 1 liter for 6 m²
SAFETY_MARGIN = 1.10   # 10% more paint
CAN_SIZE = 18
CAN_PRICE = 80.00
GALLON_SIZE = 3.6
GALLON_PRICE = 25.00


area_to_paint_in_m2 = input("Enter the area to paint in m²: ")

try:
    area_to_paint_in_m2 = float(area_to_paint_in_m2)
    liters_needed = area_to_paint_in_m2 * LITERS_PER_M2 * SAFETY_MARGIN

    # Option 1: Only 18-liter cans
    cans_only = math.ceil(liters_needed / CAN_SIZE)
    price_cans_only = cans_only * CAN_PRICE

    # Option 2: Only 3.6-liter gallons
    gallons_only = math.ceil(liters_needed / GALLON_SIZE)
    price_gallons_only = gallons_only * GALLON_PRICE

    # Option 3: Mix cans and gallons
    full_cans = int(liters_needed // CAN_SIZE)
    remaining_liters = liters_needed - (full_cans * CAN_SIZE)
    gallons_needed = math.ceil(remaining_liters / GALLON_SIZE)
    mixed_price = (full_cans * CAN_PRICE) + (gallons_needed * GALLON_PRICE)

    # Output
    print(f"\nTotal area (with 10% margin): {area_to_paint_in_m2 * SAFETY_MARGIN:.2f} m²")
    print(f"Total liters needed: {liters_needed:.2f} L\n")

    print("1. Only 18-liter cans:")
    print(f"   Cans needed: {cans_only}")
    print(f"   Total price: R$ {price_cans_only:.2f}\n")

    print("2. Only 3.6-liter gallons:")
    print(f"   Gallons needed: {gallons_only}")
    print(f"   Total price: R$ {price_gallons_only:.2f}\n")

    print("3. Optimal mix of cans and gallons:")
    print(f"   Cans: {full_cans}")
    print(f"   Gallons: {gallons_needed}")
    print(f"   Total price: R$ {mixed_price:.2f}")

except ValueError:
    print("Invalid input. Please enter a valid number for the area to paint.")
````

### 18. Download Time Estimation
Write a program that asks for:
- The **file size** in MB.
- The **internet speed** in Mbps.

Calculate and display the **approximate download time** in **minutes**.



````python
try:
    file_size_mb = float(input("Enter the file size in MB: "))
    internet_speed_mbps = float(input("Enter the internet speed in Mbps: "))

    if file_size_mb <= 0 or internet_speed_mbps <= 0:
        print("Please enter positive numbers.")
    else:
        file_size_megabits = file_size_mb * 8
        time_seconds = file_size_megabits / internet_speed_mbps
        time_minutes = time_seconds / 60

        print(f"\nApproximate download time: {time_minutes:.2f} minutes")
except ValueError:
    print("Invalid input. Please enter numeric values.")
````
