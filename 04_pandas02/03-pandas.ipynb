# Dataframe Joins

Merge or join operations combine datasets by linking rows using one or more keys. These algorithms are very common and important in relational databases. In Pandas, the `.merge` function is the main entry point for using these types of algorithms with the data stored in DataFrames.

Let's see an example:

````python
# First, let's create two dataframes.

df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7))})

df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3))})

# Now let's apply the merge function.
display(pd.merge(df1, df2, on="key"))

# Can you explain the merge table below?
# Tip: Print df1 and df2 to help with your explanation.

display(df1)
display(df2)
````

By default, `pandas.merge` performs an "inner" join; the keys in the result are the **intersection** of both tables.

Therefore, the `pd.merge(df1, df2, on="key")` instructs Pandas to perform a SQL-like join operation on `df1` and `df2` using the "key" column as the basis for the merge. The default merge operation is an inner join, which means the result will include only those rows that have matching values in the "key" column in both `df1` and `df2`.

````python
# If the column names to join on are different in each dataframe, you can specify them separately.

df3 = pd.DataFrame({"lkey": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7))})

df4 = pd.DataFrame({"rkey": ["a", "b", "d"],
                    "data2": pd.Series(range(3))})

pd.merge(df3, df4, left_on="lkey", right_on="rkey")

# Question: Where are the "c" and "d" data?
````

Other possible options are "left", "right", and "outer". The outer join takes the union of the keys, combining the effect of applying both left and right joins.

Option       | Behaviour
-------------|-----------------------------------------------
how="inner"  | Use only the key combinations observed in both tables
how="left"   | Use all key combinations found in the left table
how="right"  | Use all key combinations found in the right table
how="outer"  | Use all key combinations observed in both tables together

````python
pd.merge(df1, df2, how="outer")
````

````python
pd.merge(df3, df4, left_on="lkey", right_on="rkey", how="outer")
````

# Pandas `apply()`

The `apply()` method is one of the most common methods for data preprocessing. It simplifies applying a function to each element in a Series and to each row (or column) in a DataFrame.

````python
# Apply a function to a Series. Let's revisit the example, such as calculating the average
# distance of the planets from the sun (in million kilometers).
# Let's say that we need to convert that distance to miles. How can we do that?

distances = {
    'Mercury': 57.9,
    'Venus': 108.2,
    'Earth': 149.6,
    'Mars': 227.9,
    'Jupiter': 778.5
}
s_distances_km = pd.Series(distances)
print(s_distances_km)
````

````python
# By using apply, we can define a function to do the conversion and then "apply" this function to the entire Series.

def kilometers_to_miles(kilometers):
    miles_per_kilometer = 0.621371
    return kilometers * miles_per_kilometer

s_distances_mi = s_distances_km.apply(kilometers_to_miles)
print(s_distances_mi)
````

The code above displays the contents of the `s_distances_mi` object and its data type.

The distances are converted to miles. To do so, we first define a function that performs the conversion, then pass the **function name without parentheses** to the `apply()` method. The `apply()` method takes each element in the Series and applies the `kilometers_to_miles()` function to it.

# Pandas GroupBy

Pandas' `groupby` is a powerful feature in the Pandas library that allows for grouping data based on some criteria and then applying a function to each group independently. By using `groupby`, you can split your data into separate groups to perform computations for better analysis. For example, you might group a dataset by a category like gender or country, and then calculate the average age or total sales for each group. This method is instrumental in data exploration and preprocessing, making it easier to understand patterns and insights within different segments of your dataset.

# The 3 Steps of a GroupBy Process

Any `groupby` process involves some combination of the following three steps:

- Splitting the original object into groups based on the defined criteria.
- Applying a function to each group independently.
- Combining the results back together.

Let's explore this split-apply-combine chain step-by-step with an example from the Prize Dataset: file `prize_dataset.csv`.

````python
# First, let's load the dataset

df = pd.read_csv('prize_dataset.csv')

# TODO: Filter the dataset by the following columns: 
# 'awardYear', 'category', 'prizeAmount', 'prizeAmountAdjusted', 'name', 'gender', 'birth_continent'
````

#### Splitting the Original Object into Groups

Now, let's use the `DataFrame.groupby()` function to split the data into groups. We can do this along rows (by default, `axis=0`) or columns (`axis=1`).

For example, in our case, we can group the Nobel Prize data by prize category:

````python
grouped = df.groupby('category')
# If we try to print one of the GroupBy objects that we created, we will actually see no groups:
print(grouped)
````

It is important to note that creating a GroupBy object only checks if we have passed a correct mapping; it doesn't actually perform any operations of the split-apply-combine chain until we explicitly use some method on this object or extract some of its attributes.

````python
# To briefly inspect the resulting GroupBy object and check how the groups were exactly split, 
# we can extract the 'groups' or 'indices' attribute from it.

print(grouped.groups)
print(grouped.indices)

# Both attributes return a dictionary where the keys are the created groups, and the values are 
# lists of axis labels (for the 'groups' attribute) or indices (for the 'indices' attribute) 
# corresponding to the instances of each group in the original DataFrame:
````

````python
# To find the number of groups in a GroupBy object, we can use the 'ngroups' attribute 
# or call the 'len' function from the Python standard library.
print(grouped.ngroups)
print(len(grouped))
````

````python
# If we need to visualize all or some entries of each group, we can iterate over the GroupBy object:

# Each entry is a DataFrame!
for name, entry in grouped:
    print(f"Group: {name}")
    display(entry.head(2))  # Displays the first 2 entries of each group
````

````python
# If instead we want to select a single group in the form of a DataFrame,
# we can use the method get_group().

grouped.get_group('Literature')
````

# Applying a Function by Group

After splitting the original dataset, we can perform one of the following operations:

- **Aggregation**: Calculating a summary statistic for each group (e.g., group sizes, means, medians, or sums) and outputting a single number for many data points.
- **Transformation**: Conducting some operations by group, such as calculating the z-score for each group.
- **Filtration**: Rejecting some groups based on a predefined condition, such as group size, mean, median, or sum. This can also include filtering out particular rows from each group.

**Note: We will only focus on aggregation in this class.**

## Aggregation

To aggregate the data of a GroupBy object (i.e., to calculate a summary statistic by group), we can use the `agg()` method.

````python
grouped[['prizeAmount']].agg(["mean"])

# What are the other operations we could pass to agg?
````

## GroupBy Methods

Instead of using the `agg()` method, we can apply the corresponding pandas methods directly on a GroupBy object. 

The most common methods include:

- **Central Tendency Measures:**
  - `mean()`: Calculates the average value in each group.
  - `median()`: Finds the middle value in each group.
  - `mode()`: Identifies the most frequently occurring value(s) in each group.

- **Aggregation Functions:**
  - `sum()`: Adds up all numerical values in each group.

- **Counting Functions:**
  - `size()`: Counts the total number of entries in each group, including null values.
  - `count()`: Counts non-null entries in each group.

- **Range Functions:**
  - `min()`: Determines the smallest value in each group.
  - `max()`: Determines the largest value in each group.

- **Variability Measures:**
  - `std()`: Computes the standard deviation of each group, measuring the dispersion of the group's values.
  - `var()`: Computes the variance of each group, quantifying the spread of the group's values.

- **Uniqueness Function:**
  - `nunique()`: Provides the number of unique values in each group.

- **Descriptive Statistics:**
  - **`describe()`**: Outputs a summary of descriptive statistics including count, mean, std, min, 25%, 50%, 75%, and max values by group.
